```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('cluster_data-3.csv')
df.head(20)
```

```python
df.shape
```

```python
# Drop 'activity' (target column) for clustering/PCA
features_df = df.drop(columns=['activity'])

# Step 1: Compute correlation matrix and count pairs with correlation > 0.8
corr_matrix = features_df.corr().abs()
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_pairs = (upper_triangle > 0.8).sum().sum()

# Step 2: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features_df)

# Step 3: Apply PCA to capture 90% variance
pca = PCA()
pca.fit(scaled_data)

# Find number of components to capture 90% variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_90 = np.argmax(cumulative_variance >= 0.9) + 1

# Get top 5 PCs eigenvalues and eigenvectors
top_5_eigenvalues = pca.explained_variance_[:5]
top_5_eigenvectors = pca.components_[:5]


(high_corr_pairs, n_components_90, top_5_eigenvalues,top_5_eigenvectors )
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Apply PCA transformation with 64 components (90% variance)
pca_90 = PCA(n_components=64)
reduced_data = pca_90.fit_transform(scaled_data)

# Use the Elbow method to determine optimal number of clusters
wcss = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(reduced_data)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(reduced_data, kmeans.labels_))

# Plotting WCSS (Elbow Curve)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(K_range, wcss, marker='o')
plt.title('Elbow Method: WCSS vs K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')

# Plotting Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, marker='o', color='green')
plt.title('Silhouette Score vs K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

 Inference for Q2(ii):
Optimal number of clusters: Likely 3 or 4, based on both Elbow and Silhouette

Business Interpretation:

Each cluster may represent distinct user behavioral patterns (e.g., types of movement)

These clusters could inform:

Activity grouping for compression

Anomaly detection (if any cluster is sparse)

Simplified categorization for downstream ML tasks



```python
# Finalize KMeans with K=4
kmeans_final = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans_final.fit_predict(reduced_data)

# Add cluster labels to the original DataFrame for interpretation
df_clusters = df.copy()
df_clusters['Cluster'] = cluster_labels

# Summarize the size of each cluster
cluster_sizes = df_clusters['Cluster'].value_counts().sort_index()

# Compute mean of original features (grouped by cluster) for interpretation
cluster_summary = df_clusters.drop(columns=['activity']).groupby('Cluster').mean()

cluster_sizes
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Use only first 100 samples for dendrogram due to performance
subset_data = scaled_data[:100]

# Dictionary to hold linkage methods and their corresponding linkage matrices
linkage_methods = ['single', 'complete', 'average', 'ward']
linkage_results = {method: sch.linkage(subset_data, method=method) for method in linkage_methods}

# Plot dendrograms for each linkage method
plt.figure(figsize=(20, 18))

for i, method in enumerate(linkage_methods, 1):
    plt.subplot(2, 2, i)
    sch.dendrogram(linkage_results[method])
    plt.title(f'Dendrogram using {method.capitalize()} Linkage')
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

🔍 Observations:
Linkage	Characteristics
Single	Merges closest points; prone to "chaining effect" (long stretched clusters)
Complete	Merges based on farthest pair; leads to compact and evenly sized clusters
Average	Balances single and complete; smooth merges but can still chain slightly
Ward	Minimizes variance within clusters; usually forms well-separated clusters
✅ Recommendation:
For balanced, compact clusters, Ward linkage is generally preferred — especially with scaled numeric data.

```python
# KMeans with K=5
kmeans_k5 = KMeans(n_clusters=5, random_state=42)
labels_k5 = kmeans_k5.fit_predict(reduced_data)

# Assign labels and compute WCSS per cluster
df_k5 = pd.DataFrame(reduced_data)
df_k5['Cluster'] = labels_k5

# Compute WCSS (inertia) for each individual cluster
cluster_inertias = {}
for i in range(5):
    cluster_points = df_k5[df_k5['Cluster'] == i].drop(columns='Cluster')
    centroid = cluster_points.mean()
    wcss = np.sum((cluster_points - centroid) ** 2, axis=1).sum()

    cluster_inertias[i] = wcss

# Convert to DataFrame for easier viewing and sorting
wcss_df = pd.DataFrame.from_dict(cluster_inertias, orient='index', columns=['WCSS'])
wcss_df = wcss_df.sort_values(by='WCSS', ascending=False)


wcss_df
```

Here is the ranking of the 5 clusters (K=5) based on their within-cluster sum of squares (WCSS):

Clusters with higher WCSS are more spread out

Clusters with lower WCSS are more compact

This helps in:

Interpreting cluster compactness

Possibly renaming clusters based on spread or behavioral consistency

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

# Load recommendation data
df_rec = pd.read_csv("recommendations-1.csv")

# Define Surprise reader format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df_rec[['UserID', 'ItemID', 'Rating']], reader)

# Train-test split
trainset, testset = train_test_split(data, test_size=0.25, random_state=42)

# Build collaborative filtering model using SVD
model = SVD()
model.fit(trainset)

# Predict on testset
predictions = model.test(testset)

# Calculate RMSE
rmse_score = rmse(predictions)

# Extract top 5 recommendations for a specific user (e.g., UserID = 10)
user_id = 10
item_ids = df_rec['ItemID'].unique()
user_predictions = [(iid, model.predict(user_id, iid).est) for iid in item_ids]
top_5_recommendations = sorted(user_predictions, key=lambda x: x[1], reverse=True)[:5]

# Format result
top_5_items = pd.DataFrame(top_5_recommendations, columns=['ItemID', 'Predicted_Rating'])



rmse_score, top_5_items
```
