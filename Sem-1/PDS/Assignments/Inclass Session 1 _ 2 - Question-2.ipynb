```python
# Write your code here
my_list = [15, 78, 10, 45, 89]

def swap_first_last(my_list):
    if len(my_list) >= 3:
        my_list[0], my_list[-1] = my_list[-1], my_list[0]
        return my_list
    else:
        return my_list[::-1]

print(swap_first_last(my_list))
print(swap_first_last([12,32]))
```
