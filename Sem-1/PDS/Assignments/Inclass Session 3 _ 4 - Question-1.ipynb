**Use the matrix given below:**

```python
# type your code here
M = [[15, 17, 45, 56], [7, 42, 15, 63], [54, 3, 61, 41], [0, 87, 16, 20]]

def print_matrix(arr):
  for item in arr:
    print(item)

# method 1: using map and zip helpers
transpose_M = list(map(list, zip(*M)))

print("Transpose of matrix with map and zip helpers: ")
print_matrix(transpose_M)

# method 2: using loops by swapping values
def transpose(A):
  print("Before transpose: ")
  print_matrix(A)
  for i in range(0, len(A)):
    for j in range(i, len(A)):
      t = A[i][j]
      A[i][j] = A[j][i]
      A[j][i] = t
  print("After transpose: ")
  print_matrix(A)

transpose(M)
```
