```python
opening_bal = []
closing_bal = []
interest = []
contribution = []
deposited_amount = []
month_count = []
in_hand_sal = []
def compound_interest_with_deposit_increase(principal, annual_interest_rate, months, initial_deposit, deposit_increase_interval, deposit_increase_percentage, in_hand):
  """
  Calculates the compound interest for a given principal, annual interest rate,
  number of months, initial monthly deposit, deposit increase interval, and
  deposit increase percentage.

  Args:
    principal: The initial principal amount.
    annual_interest_rate: The annual interest rate as a decimal (e.g., 0.0825 for 8.25%).
    months: The number of months for the investment.
    initial_deposit: The initial monthly deposit amount.
    deposit_increase_interval: The number of months between deposit increases.
    deposit_increase_percentage: The percentage increase in deposit every interval.

  Returns:
    The final balance after the specified number of months.
  """

  monthly_interest_rate = annual_interest_rate / 12
  balance = principal
  monthly_deposit = initial_deposit

  for month in range(months):
    opening_bal.append(balance)
    contribution.append(monthly_deposit)
    interest_earned = balance * monthly_interest_rate
    interest.append(interest_earned)
    deposited_amount.append(monthly_deposit + interest_earned)
    month_count.append(month + 1)
    balance += monthly_deposit + interest_earned

    closing_bal.append(balance)
    in_hand_sal.append(in_hand)
    if (month + 1) % deposit_increase_interval == 0:
      monthly_deposit *= (1 + deposit_increase_percentage)
      in_hand *= ( 1 + deposit_increase_percentage)

    if (month + 1) % deposit_increase_interval != 0 and (month + 1) % 12 == 0:
      in_hand *= (1.04)


  return balance

# Given values
principal = 422151
annual_interest_rate = 0.0825
months = 20 * 12  # 20 years in months
initial_deposit = 11902
deposit_increase_interval = 3 * 12  # Increase deposit every 3 years
deposit_increase_percentage = 0.40  # 40% increase
in_hand = 79000

# Calculate the closing balance
closing_balance = compound_interest_with_deposit_increase(principal, annual_interest_rate, months, initial_deposit,
                                                          deposit_increase_interval, deposit_increase_percentage,
                                                          in_hand)

print(f"Closing balance after twenty years: {closing_balance:.2f}")
```

```python
import pandas as pd

def calculate_loan_statement(loan_amount, annual_rate, loan_tenure, part_payment=0):
    # Convert annual interest rate to monthly rate
    monthly_rate = annual_rate / 100 / 12

    # Calculate EMI using the formula
    emi = loan_amount * (monthly_rate * (1 + monthly_rate) ** loan_tenure) / ((1 + monthly_rate) ** loan_tenure - 1)

    # Initialize variables
    principal_remaining = loan_amount
    loan_data = []

    # Generate the loan statement for each month
    for month in range(1, loan_tenure + 1):
        # Calculate interest for this month
        interest_paid = principal_remaining * monthly_rate
        # Calculate principal cleared in this month
        principal_cleared = emi - interest_paid + part_payment
        # Remaining principal
        principal_remaining -= principal_cleared

        # Ensure that the last month's principal remaining is 0
        if principal_remaining < 0:
            principal_remaining = 0

        # Store the values in the list
        loan_data.append({
            'Month': month,
            'EMI_Amount': round(emi, 2),
            'Interest_Paid': round(interest_paid, 2),
            'Principal_Cleared': round(principal_cleared, 2),
            'Principal_Remaining': round(principal_remaining, 2),
            'Part_Payment': part_payment
        })

    # Create a DataFrame from the loan data
    df = pd.DataFrame(loan_data)

    return df

# Input data (Example)
loan_amount = 5000000   # Loan amount in INR
annual_rate = 8         # Annual interest rate in percentage
loan_tenure = 240       # Loan tenure in months (20 Years)

# Get the loan statement
loan_statement_df = calculate_loan_statement(loan_amount, annual_rate, loan_tenure)

# Display the DataFrame
loan_statement_df.head()
```
