# Aggregation


```python
import numpy as np

np.random.seed(23) # seeding random generator to follow / give same array every time instead of random values
n64 = np.random.randint(low=1, high=20, size=(6,4))

print(n64)
print(np.sum(n64))
print(np.sum(n64, axis=0)) # this will add columns and gives 1D array
print(np.sum(n64, axis=1)) # this will give 1D array by adding rows into one element 
```

```python
np.random.seed(23)
n89 = np.random.randint(low=10, high=100, size=(8,9))
n89
```

```python
print(np.vsplit(n89, 2)) # split array into two eqaul sizes vertically
print(np.hsplit(n89, 3)) # vice versa horizontally
```

```python
# splitting the array using indices
print(np.vsplit(n89, [3,5,7])) # 0:3:5:7 vertical slicing
print(np.hsplit(n89, [2,4,5,7])) # 0:2:4:5:7 horizontal slicing
```

## Concat / Stack


```python
n34 = np.arange(12).reshape(3,4)
n36 = np.arange(100, 118).reshape(3,6)
n64 = np.arange(200, 224).reshape(6,4)

print(np.concatenate((n34, n64), axis=0).shape) # This will add two arrays vertically if they have same size of columns
# this can be achived with np.vstack(n34,n64)
print(np.concatenate((n34, n36), axis=1).shape)  # This will add two arrays vertically if they have same size of rows
# this can be achived with np.hstack(n34, n36)
```

# Mapping Data Frame with Dictionary values

```python
md = { "Hyd": 'SZ', "Blr": "SZ", "Che": "SZ", "Mum": "WZ", "NCR": "NZ", "Kol": "EZ"}

df["Zone"] = df.HQ.map(md)

df
```

```python
def sales(vol_value):
    if vol_value > df.Vol.mean():
        return "Good Sales"
    else:
        return "Average Sales"
df["sales_performance"] = df.Vol.map(sales)
df
```

```python
df["sales_performance"] = df.Vol.map(lambda v: "Good Sales" if v > df.Vol.mean() else "Average Sales")
df
```

## Pivot Table

```python
pd.pivot_table(df1, index="Sector")
```

```python
pd.pivot_table(df1, index="Sector", values="Rev", aggfunc="mean")
```

```python
pd.pivot_table(df1, index=["Sector", "GST"], values=["Rev", "Vol"], aggfunc=["mean", "min", "max", lambda col: col.max() - col.min() ])
```

```python
diff = lambda col: col.max() - col.min()
diff.__name__ = "Difference"
pd.pivot_table(df1, index=["Sector", "GST"], values=["Rev", "Vol"], aggfunc=["mean", "min", "max",  diff])
```
