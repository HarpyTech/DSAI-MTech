* Type of Test: We will use a two-sample t-test assuming equal variances to compare the means of the two catalysts.

#### ii. Test the hypothesis and conclude
  To test the hypothesis, we will perform a two-sample t-test. Here are the steps:

* Calculate the means and standard deviations of the two samples.

- Calculate the pooled standard deviation.

- Calculate the t-statistic.

- Determine the critical t-value for a two-tailed test with

```python
# Given data
catalyst1 = np.array([91.50, 94.18, 92.18, 95.39, 91.79, 89.07, 94.72, 89.21])
catalyst2 = np.array([89.19, 90.95, 90.46, 93.21, 97.19, 97.04, 91.07, 92.75])

# Calculate means
mean1 = np.mean(catalyst1)
mean2 = np.mean(catalyst2)

# Calculate standard deviations
std1 = np.std(catalyst1, ddof=1)
std2 = np.std(catalyst2, ddof=1)

# Calculate pooled standard deviation
n1 = len(catalyst1)
n2 = len(catalyst2)
pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

# Calculate t-statistic
t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))

# Degrees of freedom
df = n1 + n2 - 2

# Critical t-value for two-tailed test with alpha = 0.05
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

# Calculate p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Print results
print(f"Mean of Catalyst 1: {mean1:.2f}")
print(f"Mean of Catalyst 2: {mean2:.2f}")
print(f"Standard Deviation of Catalyst 1: {std1:.2f}")
print(f"Standard Deviation of Catalyst 2: {std2:.2f}")
print(f"Pooled Standard Deviation: {pooled_std:.2f}")
print(f"t-statistic: {t_statistic:.2f}")
print(f"Critical t-value: {t_critical:.2f}")
print(f"p-value: {p_value:.4f}")

# Conclusion
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: There is a significant difference in the mean yields of the two catalysts.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean yields of the two catalysts.")
```

Type of Test: We will use a one-sample t-test to compare the sample mean to the known population mean.

ii. Test the hypothesis and conclude

To test the hypothesis, we will perform a one-sample t-test. Here are the steps:

Calculate the sample mean and standard deviation.

Calculate the standard error of the mean.

Calculate the t-statistic.

Determine the critical t-value for a two-tailed test with

```python
# Given data
population_mean = 14.35
population_std = 9.73
sample_mean = 11.77
sample_size = 30
alpha = 0.05

# Calculate the standard error of the mean
standard_error = population_std / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Degrees of freedom
df = sample_size - 1

# Critical t-value for two-tailed test with alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

# Calculate p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Mean: {population_mean:.2f}")
print(f"Standard Deviation: {population_std:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"t-statistic: {t_statistic:.2f}")
print(f"Critical t-value: {t_critical:.2f}")
print(f"p-value: {p_value:.4f}")

# Conclusion
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: There is a significant change in the mean P/E ratio of NYSE stocks.")
else:
    print("Fail to reject the null hypothesis: There is no significant change in the mean P/E ratio of NYSE stocks.")
```

#### 3d) General hospitals patient account division has compiled data on the age of accounts receivable. The data collected indicate that the age of the accounts follows a normal distribution with μ = 28 days and σ = 8 days.

- What portion of the accounts is between 20 and 40 days old. (4 mark)

- The hospital administrator is interested in sending reminder letters to the oldest 15% of accounts. How many days old should an account be, before a reminder letter is sent. (2 mark)

```python
# Given data
mean = 28
std_dev = 8

# Calculate the CDF for 20 and 40 days
cdf_20 = stats.norm.cdf(20, mean, std_dev)
cdf_40 = stats.norm.cdf(40, mean, std_dev)

# Calculate the portion of accounts between 20 and 40 days old
portion_between_20_and_40 = cdf_40 - cdf_20

print(f"Portion of accounts between 20 and 40 days old: {portion_between_20_and_40:.4f}")

# Calculate the 85th percentile
percentile_85 = stats.norm.ppf(0.85, mean, std_dev)

print(f"Accounts should be {percentile_85:.2f} days old before a reminder letter is sent.")
```

### 3e) A pharmaceutical company claims that its new tablet is effective in increasing the height of children. The data of heights (in cm) of 7 children is recorded before and after consuming the tablet.

- Check for normality of the data and perform appropriate test.

- Test the company's claim at a 5% level of significance using the p-value approach.


    ht_before = [121, 125, 130, 120, 145, 126, 134]
    
    ht_after = [130, 129, 148, 122, 147, 130, 148]

### Step 1: Check for Normality
  We will use the Shapiro-Wilk test to check for normality of the data. Additionally, we can visualize the data using Q-Q plots.

### Step 2: Perform the Appropriate Test
  
  Since we have paired data (heights before and after consuming the tablet), we will use a paired t-test to compare the means of the two related groups.

```python
# Given data
ht_before = [121, 125, 130, 120, 145, 126, 134]
ht_after = [130, 129, 148, 122, 147, 130, 148]

# Check for normality using Shapiro-Wilk test
shapiro_before = stats.shapiro(ht_before)
shapiro_after = stats.shapiro(ht_after)

print(f"Shapiro-Wilk test for heights before: W={shapiro_before[0]:.4f}, p-value={shapiro_before[1]:.4f}")
print(f"Shapiro-Wilk test for heights after: W={shapiro_after[0]:.4f}, p-value={shapiro_after[1]:.4f}")

# Plot Q-Q plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
stats.probplot(ht_before, dist="norm", plot=plt)
plt.title('Q-Q Plot for Heights Before')

plt.subplot(1, 2, 2)
stats.probplot(ht_after, dist="norm", plot=plt)
plt.title('Q-Q Plot for Heights After')

plt.tight_layout()
plt.show()

# Perform paired t-test
t_statistic, p_value = stats.ttest_rel(ht_before, ht_after)

print(f"Paired t-test: t-statistic={t_statistic:.4f}, p-value={p_value:.4f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The tablet is effective in increasing the height of children.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to support the claim that the tablet is effective in increasing the height of children.")
```

Explanation
Check for Normality:

* Shapiro-Wilk Test: This test checks if the data is normally distributed. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the data is normally distributed.

* Q-Q Plots: These plots help visualize if the data follows a normal distribution. If the points lie along the reference line, the data is normally distributed.

Perform Paired t-Test:

* Paired t-Test: This test compares the means of two related groups. We calculate the t-statistic and p-value to determine if there is a significant difference between the heights before and after consuming the tablet.

Conclusion:

* Based on the p-value and the significance level (
α
= 0.05
), we decide whether to reject or fail to reject the null hypothesis.

Results
* Normality Check: The Shapiro-Wilk test results and Q-Q plots will indicate if the data is normally distributed.

- Paired t-Test: The t-statistic and p-value will help us determine if there is a significant increase in height after consuming the tablet.

### 4b) Consider the insurance.csv file and answer the following questions

- i)	Check whether the BMI data follows normal distribution by using a qq plot and shaipro test ( 3marks)

- ii)	Construct a hypothesis to prove that smoker have higher charges than non smoker and conclude accordingly ( 4 marks)

- iii)	Is there a statisitical difference between insurance charges for male and female and conclude accordingly ( 4 marks)

- iv)	Is the proportion of smokers significantly different in different genders? (4 Marks)


```python
# Load the dataset
df = pd.read_csv('insurance.csv')

# Extract the BMI data
bmi = df['bmi']

# Q-Q Plot
plt.figure(figsize=(8, 6))
stats.probplot(bmi, dist="norm", plot=plt)
plt.title('Q-Q Plot of BMI')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()

# Shapiro-Wilk Test
shapiro_test = stats.shapiro(bmi)
print(f'Shapiro-Wilk Test: W={shapiro_test[0]:.4f}, p-value={shapiro_test[1]:.4f}')
```

**Shapiro-Wilk Test**: This statistical test checks for normality. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the data is normally distributed.

###Question 2
#### Hypothesis:

- Null Hypothesis (H₀): There is no difference in charges between smokers and non-smokers.

- Alternative Hypothesis (H₁): Smokers have higher charges than non-smokers.

```python
# Extract charges for smokers and non-smokers
smokers_charges = df[df['smoker'] == 'yes']['charges']
non_smokers_charges = df[df['smoker'] == 'no']['charges']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(smokers_charges, non_smokers_charges, alternative='greater')

print(f'Two-sample t-test: t-statistic={t_stat:.4f}, p-value={p_value:.4f}')
```

we reject the null hypothesis and conclude that smokers have higher charges than non-smokers. **Since P-Value < 0.05**

Q3

Hypothesis:

Null Hypothesis (H₀): There is no difference in insurance charges between males and females.

Alternative Hypothesis (H₁): There is a difference in insurance charges between males and females.

Test: We can use a two-sample t-test to compare the means of the two groups (males and females).

```python
# Extract charges for males and females
male_charges = df[df['sex'] == 'male']['charges']
female_charges = df[df['sex'] == 'female']['charges']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(male_charges, female_charges)

print(f'Two-sample t-test: t-statistic={t_stat:.4f}, p-value={p_value:.4f}')
```

**p-value is less than 0.05**, we reject the null hypothesis and conclude that there is a significant difference in insurance charges between males and females.

Hypothesis:

Null Hypothesis (H₀): The proportion of smokers is the same across different genders.

Alternative Hypothesis (H₁): The proportion of smokers is different across different genders.

Test: We can use a chi-square test for independence to compare the proportions of smokers in different genders.

```python
# Create a contingency table
contingency_table = pd.crosstab(df['sex'], df['smoker'])

# Perform chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f'Chi-square test: chi2_statistic={chi2_stat:.4f}, p-value={p_value:.4f}')
```

**p-value is less than 0.05**, we reject the null hypothesis and conclude that the proportion of smokers is significantly different across genders.
